```python
#Navithanjali.M.S (24BAD077)
# Descriptive Statistics & Visualization
print("Navithanjali.M.S (24BAD077)\n")
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")

# Load Dataset
df = pd.read_csv("happiness_2017.csv")

# Basic Information
print("\nDataset Shape:", df.shape)
print("\nDataset Info:\n")
df.info()

print("\nFirst 5 Rows:\n")
print(df.head())

# Missing Values
print("\nMissing Values:\n")
print(df.isnull().sum())

# Descriptive Statistics
print("\nDescriptive Statistics:\n")
print(df.describe())

# Histogram
plt.figure()
df['HappinessScore'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title("Histogram of Happiness Score")
plt.xlabel("Happiness Score")
plt.ylabel("Frequency")
plt.show()

# Box Plot
plt.figure()
plt.boxplot(df['HappinessScore'])
plt.title("Box Plot of Happiness Score")
plt.ylabel("Happiness Score")
plt.show()

# Count Plot (Region)
plt.figure(figsize=(10,6))
sns.countplot(y='Region', data=df, palette='viridis')
plt.title("Count of Countries by Region")
plt.show()

# Line Plot
plt.figure()
plt.plot(df['Rank'], df['HappinessScore'], marker='o', linestyle='-', color='orange')
plt.xlabel("Happiness Rank")
plt.ylabel("Happiness Score")
plt.title("Happiness Rank vs Happiness Score")
plt.show()

# Bar Plot (Top 10 Happiest Countries)
top10 = df.sort_values(by='HappinessScore', ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.barh(top10['Country'], top10['HappinessScore'], color='purple')
plt.xlabel("Happiness Score")
plt.title("Top 10 Happiest Countries")
plt.gca().invert_yaxis()  # Highest happiness on top
plt.show()

# Correlation Heatmap
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()

# Pie Chart (Region Distribution)
region_counts = df['Region'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title("Region-wise Distribution of Countries")
plt.show()

#Scatter plot
plt.figure(figsize=(8,6))
plt.scatter(df['Log GDP per capita'], df['HappinessScore'], color='green', alpha=0.7, edgecolors='black')
plt.xlabel("Log GDP per Capita")
plt.ylabel("Happiness Score")
plt.title("GDP per Capita vs Happiness Score")
plt.grid(True)
plt.show()
```

Descriptive Theory for Visualizations

1. Histogram

Purpose: To show the distribution of a numeric variable.

Example in dataset: Histogram of HappinessScore.

Interpretation:

Helps identify if the happiness scores are normally distributed, skewed, or have outliers.

Peaks in the histogram indicate common happiness levels among countries.

2. Box Plot (Whisker Plot)

Purpose: To summarize data distribution using median, quartiles, and detect outliers.

Example in dataset: Box plot of HappinessScore.

Interpretation:

Shows the median happiness score and the spread of scores.

Outliers represent countries with exceptionally low or high happiness compared to the rest.

3. Count Plot / Bar Plot

Purpose: To show the frequency of categorical variables.

Example in dataset: Count of countries by Region.

Interpretation:

Identifies which regions have more countries.

Helps in comparing categorical data visually.

4. Line Plot

Purpose: To show trends across ordered data.

Example in dataset: Rank vs HappinessScore.

Interpretation:

Displays how happiness decreases with rank.

Easy to see gradual changes across countries.

5. Pie Chart

Purpose: To show the proportion of categories in a dataset.

Example in dataset: Region-wise distribution of countries.

Interpretation:

Helps understand percentage of countries in each region.

Visualizes the composition of a categorical variable.

6. Correlation Heatmap

Purpose: To visualize correlation between numeric variables.

Example in dataset: Heatmap of HappinessScore, Log GDP per capita, Social support, etc.

Interpretation:

Darker or lighter colors indicate stronger positive or negative correlations.

Highlights which factors have the highest influence on HappinessScore.

7. Scatter Plot

Purpose: To visualize relationship between two numeric variables.

Example in dataset: Log GDP per capita vs HappinessScore.

Interpretation:

Positive trend indicates higher GDP per capita is associated with higher happiness.

Detects patterns, clusters, or outliers in the data.
